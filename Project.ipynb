# Start Cleaning

```python
# Drop the incomplete surveyors from the original DataFrame
df.dropna(how='all', inplace=True)
df.tail()
```

## Time to drop a lot of columns

```python
# Reviewing the ReadMe file shows a lot of the columns are just survey questions that won't really be useful.
# Ethnicity_text isn't specified in the file but I'm guessing it was the \
# ethnicities written out instead of categorized by numbers and they just decided not to use it.
# Majority of the columns at the end are survey questions such as how \
# many hours a day do you work out between different types of exercises.
# Decided to keep the Hobbies_Time_# columns since they may be a proponent of happiness.
ss = df[['Age','Ethnicity','Gender','Year_credits','Program','Part_time','Living','Disability',
         'Province','International','Hours_sleep','Mindfulness_freq','Diagnosis','Hobbies_Time_1',
        'Hobbies_Time_2','Hobbies_Time_3','Hobbies_Time_4','Hobbies_Time_5','Hobbies_Time_6','Hobbies_Time_7','Hobbies_Time_8',]]
ss.head()
```

```python
# SS = Student Survey
# Year_credits = college year based on credit hours (e.g., freshman, sophomore, etc.).
# There was a Year_calendar, but I didn't see a need to keep both it and Year_credits.
# A lot of the column values represent yes/no or actual text.
# Check if there are any more NaN values.

ss.info()
#ss[ss.isna().any(axis=1)]
```

```python
# Check how many unique values are in each column
ss.nunique()
```

## Split Hobbies_Time

```python
# Grabs the first 4 Hobbies_Time columns.
# Combines the actual Hobby columns: things like watching movies or playing games.
# This makes it easier to visualize how many hours a week went into leisure time.
columns_to_join = ss.filter(like='Hobbies_Time').columns[:4]
all_hobbies = ss.loc[:, columns_to_join].sum(axis=1)

# Drops the columns joined above
ss = ss.drop(columns=columns_to_join).assign(Hobbies_Hours=all_hobbies)
ss
```

```python
# Copies the previous line since there are only 4 hobby columns left.
# Hobbies_Time_5-8 is time spent working, studying, or using educational resources.
# Joining these two groups of "hobbies" allows a cleaner showing of their correlational values for later.
columns_to_join = ss.filter(like='Hobbies_Time').columns[:4]
study_time = ss.loc[:, columns_to_join].sum(axis=1)

# Drops the columns joined
ss = ss.drop(columns=columns_to_join).assign(Study_Hours=study_time)
ss
```

## Replace Column Number values with Categorical 

```python
# Use map to rename the float values as their object equivalent.
# All the object values come from the ReadMe file.
# Needed to assign the change to a variable to avoid the SettingWithCopy error.

mapped_ethnicity = ss['Ethnicity'].map({
    1.0: 'Caucasian', 
    2.0: 'Asian', 
    3.0: 'Indigenous',
    4.0: 'Hispanic/Latino', 
    5.0: 'Black', 
    6.0: 'Other', 
    7.0: 'Metis'})
ss['Ethnicity'] = mapped_ethnicity


mapped_gender = ss['Gender'].map({
    1.0: 'Male', 
    2.0: 'Female', 
    3.0: 'non-binary',
    12.0: 'Other'})
ss['Gender'] = mapped_gender


mapped_year = ss['Year_credits'].map({
    1.0: 'Freshman', 
    2.0: 'Sophomore', 
    3.0: 'Junior',
    4.0: 'Senior',
    5.0: 'Super Senior'})
ss['Year_credits'] = mapped_year


mapped_program = ss['Program'].map({
    1.0: 'Business', 
    2.0: 'Education', 
    3.0: 'Engineering',
    4.0: 'Fine Arts',
    5.0: 'Human and Social Development', 
    6.0: 'Humanities', 
    7.0: 'Sciences', 
    8.0: 'Social Sciences'})
ss['Program'] = mapped_program


mapped_schedule = ss['Part_time'].map({
    1.0: 'Part-Time', 
    2.0: 'Full-Time'})
ss['Part_time'] = mapped_schedule


mapped_living = ss['Living'].map({
    1.0: 'On-campus', 
    2.0: 'Off-campus/Family', 
    3.0: 'Off-campus/Friends',
    4.0: 'Off-campus/Alone'})
ss['Living'] = mapped_living


mapped_disability = ss['Disability'].map({
    1.0: 'Yes', 
    2.0: 'No', 
    4.0: 'Prefer not to answer'})
ss['Disability'] = mapped_disability


mapped_province = ss['Province'].map({
    1.0: 'British Columbia', 
    2.0: 'Alberta', 
    3.0: 'Saskatchewan',
    4.0: 'Manitoba',
    5.0: 'Ontario', 
    6.0: 'Quebec', 
    7.0: 'New Brunswick', 
    8.0: 'Newfoundland/Labrador',
    9.0: 'Nova Scotia', 
    10.0: 'Prince Edward Island', 
    11.0: 'Yukon', 
    12.0: 'Northwest Territories',
    13.0: 'Nunavut'})
ss['Province'] = mapped_province

mapped_international = ss['International'].map({
    1.0: 'Yes', 
    2.0: 'No'})
ss['International'] = mapped_international


# Mindfulness refers to things such as meditation 
mapped_mindfulness = ss['Mindfulness_freq'].map({
    1.0: 'Never', 
    2.0: 'Rarely', 
    3.0: 'Sometimes',
    4.0: 'Often',
    5.0: 'Very Often'})
ss['Mindfulness_freq'] = mapped_mindfulness


# Mental health diagnoses
mapped_diagnosis = ss['Diagnosis'].map({
    1.0: 'Yes', 
    2.0: 'No',
    3.0: 'Prefer not to answer'})
ss['Mental_disorder'] = mapped_diagnosis
ss.drop(columns='Diagnosis', inplace=True)

ss.head()
```

```python
# Need to find data/columns that can be used to measure the rest.
# Choosing to add multiple question-based columns to see which are most relevant to the DataFrame.
# Choosing to combine the columns from the old DataFrame and just add them to the ss DataFrame.
ss['Depression_scale'] = df.filter(like='DASS').sum(axis=1)
ss['Anxiety_scale'] = df.filter(like='GAD').sum(axis=1)
ss['Sickness_scale'] = df.filter(like='PHQ').sum(axis=1)
ss['Stress_scale'] = df.filter(like='PSS').sum(axis=1)
ss['Mindfulness_scale'] = df.filter(like='CAM').sum(axis=1)
ss['Support_seeking_scale'] = df.filter(like='SSS').sum(axis=1)

# The scales are based on questions that were asked in the survey which were worth points...\
# The higher the number, the more aligned the person felt with the topic.

# Scale from 21.0 - 84.0
#ss.sort_values(by='Depression_scale', ascending=False)

# Scale from 4.0 - 16.0
#ss.sort_values(by='Anxiety_scale', ascending=False)

# Scale from 4.0 - 18.0
#ss.sort_values(by='Sickness_scale', ascending=False)

# Scale from 20.0 - 100.0
#ss.sort_values(by='Stress_scale', ascending=False)
```

# Regression Model Setup

## Make dummies

```python
# Creating dummy columns of the categorical columns
catColumns = ['Ethnicity','Gender','Year_credits','Program','Part_time','Living',
              'Disability','Province','International','Mindfulness_freq','Mental_disorder']
dummies = pd.get_dummies(ss[catColumns])
dummies.info()
```

```python
# Drops the original categorical columns, then joins the dummy columns to the new df
surveyDummies = ss.drop(columns=catColumns)
surveyDummies = surveyDummies.join(dummies)
surveyDummies.info()
```

## Rescale and find correlations

```python
from sklearn.preprocessing import StandardScaler

# Using StandardScaler to rescale the numerical columns for use when finding correlation
scaler = StandardScaler()
numCols = ['Age','Hours_sleep','Hobbies_Hours','Study_Hours']
surveyDummies[numCols] = scaler.fit_transform(surveyDummies[numCols])
surveyDummies.head()
```

```python
# Using the dummy DataFrame to find correlation for each of the "question columns."
# Looks like the question columns are best at predicting each other... :|
#surveyDummies.corr()[['Depression_scale']].sort_values(by='Depression_scale', ascending=False)
#surveyDummies.corr()[['Anxiety_scale']].sort_values(by='Anxiety_scale', ascending=False)
surveyDummies.corr()[['Sickness_scale']].sort_values(by='Sickness_scale', ascending=False)
#surveyDummies.corr()[['Stress_scale']].sort_values(by='Stress_scale', ascending=False)
```

## Select Columns for prediction

```python
# Despite pretty low correlational values from the question and categorical columns,\
# I still want to see if I can predict their values.
# Keeping the question columns for predictions and two other numerical columns to see if they help.
columns_to_keep = ['Age','Hours_sleep','Gender','Depression_scale',
                   'Anxiety_scale','Sickness_scale','Stress_scale']
surveyScaled = ss[columns_to_keep]
surveyScaled.sort_values(by='Sickness_scale', ascending=False)
```

## Start Predicting

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Age was taken out due to it skewing the predictions.
# May have another use for age but it doesn't help in predictions,\
# 18-year-olds seem to have the highest affliction pertaining to the questions asked.

# Of the different combinations of these columns, Anxiety_scale had the highest prediction rate...
# 55% is still extremely low.
x_train, x_test, y_train, y_test = train_test_split(
    ss[['Depression_scale','Sickness_scale','Stress_scale','Hours_sleep']], ss[['Anxiety_scale']],
    test_size=0.30, random_state=20)

model = LinearRegression()
model.fit(x_train, y_train)

model.score(x_test,y_test)
```

```python
# As expected, the training data has a lower score than the test data.
model.score(x_train, y_train)
```

```python
y_predicted = model.predict(x_test)
```

```python
# create DataFrame for the anxiety predictions
predicted = pd.DataFrame(y_predicted, columns=['predictedAnxiety'])

# Combine the test data and the predicted data into a frame.
# The predictions align to only being right 50% of the time,
# however it is not reliable for future predictions.
final = predicted.join([x_test.reset_index(drop=True), 
                        y_test.reset_index(drop=True)])
final[['Depression_scale','Sickness_scale','Stress_scale','Hours_sleep','Anxiety_scale','predictedAnxiety']].head()
```

## Visualize Prediction Accuracy

```python
# Can further observe the accuracy of the predictions using residuals
final['residual'] = final['Anxiety_scale'] - final['predictedAnxiety']
final.head()
```

```python
# Visual representation of the accuracy of the predictions
sns.kdeplot(data=final, x='residual')
```

```python
# Using cross-validation to show exactly how much the predictors explain the dependent variable.
# Since this was a survey, variability between each participant is expected.
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression


predictor_columns = ['Depression_scale', 'Sickness_scale', 'Stress_scale','Hours_sleep']
X = final[predictor_columns]
y = final['Anxiety_scale']                             

# Create the model
model = LinearRegression()

# Perform cross-validation
# Creates "groups" to cross-examine how well the chosen independent variables predict the dependent
scores = cross_val_score(model, X, y, cv=5, scoring='r2')

# 51% aligns with the scoring the model received, 
# meaning ~49% of the variability in the dependent variable is not explained by the features
print("Cross-validated R²:", scores.mean())
```
